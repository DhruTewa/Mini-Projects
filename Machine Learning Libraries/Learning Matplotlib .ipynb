## Matplot Example workflow

```python
## 1st step - import matplotlib and get it read

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 2nd step - prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 3rd step - Setup plot
fig, ax=plt.subplots(figsize=(5,5))

# 4th step - plot data
ax.plot(x,y)

#5th step - customize the plot
ax.set(title='Simple plot',
       xlabel='x-axis',
      ylabel='y-axis')

#6th step - show plot
fig.show;
```

## Making plots using numpy

We will create:

* Line Plot
* Scatter PLot
* Barplot
* Histogram
* Subplot

```python
# Creating data
x=np.linspace(0,10,100)
```

```python
# Create line plot
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
# Create scatter plot
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Create scatter plot
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Creating bar plot using dictioinary

nut_butter_prices={"Almond Butter":10,
                  "Peanut Butter":6,
                  "Cashew Butter":12}
fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Nut Butter Prices",ylabel="Prices ($)");
```

```python
nut_butter_prices={"Almond Butter":10,
                  "Peanut Butter":6,
                  "Cashew Butter":12}
fig,ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax.set(title="Nut Butter Prices",xlabel="Prices ($)");
```

```python
# Creating the histogram
x=np.random.randn(1000)
fig, ax=plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# 1st Option
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,10))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# 2nd Option
fig, ax=plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,10))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Plotting from Pandas Dataframe

```python
car_sales_data=pd.read_csv('car-sales.csv')
car_sales_data
```

```python
car_sales_data['Price']=car_sales_data['Price'].str.replace('[\$\,\.]','')
car_sales_data
```

```python
## Removing the last two zero
car_sales_data['Price']=car_sales_data['Price'].str[:-2]
car_sales_data
```

```python
car_sales_data["Sale Date"]=pd.date_range('01/01/2020',periods=len(car_sales_data))
car_sales_data
```

```python
car_sales_data['Total_Sales']=car_sales_data['Price'].astype(int).cumsum()
car_sales_data
```

```python
# Plotting total sales
car_sales_data.plot(x='Sale Date',y='Total_Sales');
```

```python
# plotting the scatter plots
car_sales_data['Price']=car_sales_data['Price'].astype(int)
car_sales_data.plot(x='Odometer (KM)',y='Price',kind="scatter");
```

```python
#plotting the bar chart
car_sales_data.plot(x='Make',y='Odometer (KM)',kind='bar')
```

```python
#plotting histogram
car_sales_data['Odometer (KM)'].plot.hist();
```

```python
# Creating plots using OO method
heart_data=pd.read_csv('heart-disease.csv')
```

```python
heart_data.head()
```

```python
# Selecting the data with age >50
hd_over50=heart_data[heart_data['age']>50]
```

```python
hd_over50.head()
len(hd_over50)
```

```python
# Creating scatter plot for age & cholestrol w.r.t. target
fig, ax=plt.subplots(figsize=(10,6))
#Plot the data 
scatter=ax.scatter(x=hd_over50['age'],
                  y=hd_over50['chol'],
                  c=hd_over50['target'])
# customize the chart
ax.set(title='Heart vs Cholestrol',
      xlabel='Age',
      ylabel='Cholestrol')
# Adding legend to the plot
ax.legend(*scatter.legend_elements(),title='target');

# Add horizontal line
ax.axhline(hd_over50['chol'].mean(),
          linestyle='--');
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)
#Plot the data 
scatter=ax0.scatter(x=hd_over50['age'],
                  y=hd_over50['chol'],
                  c=hd_over50['target'])
# customize the chart
ax0.set(title='Heart vs Cholestrol',
            ylabel='Cholestrol')
# Adding legend to the plot
ax0.legend(*scatter.legend_elements(),title='target');

# Add horizontal line
ax0.axhline(hd_over50['chol'].mean(),
          linestyle='--');
#Plot the data 
scatter=ax1.scatter(x=hd_over50['age'],
                  y=hd_over50['thalach'],
                  c=hd_over50['target'])
# customize the chart
ax1.set(title='Heart vs Heart Rate',
        xlabel='Age',
            ylabel='Heart Rate')
# Adding legend to the plot
ax1.legend(*scatter.legend_elements(),title='target');

# Add horizontal line
ax1.axhline(hd_over50['thalach'].mean(),
          linestyle='--');

#Adding title to plot

fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold');

```
